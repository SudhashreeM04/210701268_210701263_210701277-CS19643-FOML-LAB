```python
import tkinter as tk
from tkinter import scrolledtext, messagebox, END, Label, Entry, Button, OptionMenu, StringVar
from tkinter.scrolledtext import ScrolledText
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
from collections import Counter
from heapq import nlargest

# Load English tokenizer, tagger, parser, NER and word vectors
nlp = spacy.load("en_core_web_sm")

root = tk.Tk()

root.title("Text Summarizer")

root.geometry("800x600")


root.configure(bg="lightblue")

# Function to summarize the text
def summarize_text():
    original_text = original_text_box.get("1.0", END)
    summarized_text = summarize(original_text)
    summarized_text_box.delete("1.0", END)
    summarized_text_box.insert("1.0", summarized_text)

# Function to perform text summarization
def summarize(text):
    doc = nlp(text)
    tokens = [token.text.lower() for token in doc if token.text.lower() not in STOP_WORDS and token.text not in punctuation]
    word_freq = Counter(tokens)
    max_freq = max(word_freq.values())
    for word in word_freq.keys():
        word_freq[word] = word_freq[word]/max_freq
    sent_scores = {}
    for sent in doc.sents:
        for word in sent:
            if word.text.lower() in word_freq.keys():
                if sent not in sent_scores.keys():
                    sent_scores[sent] = word_freq[word.text.lower()]
                else:
                    sent_scores[sent] += word_freq[word.text.lower()]
    summary_sentences = nlargest(3, sent_scores, key=sent_scores.get)
    summarized_text = ' '.join([sent.text for sent in summary_sentences])
    return summarized_text

label = Label(root, text="Original Text", font=("Arial", 14), bg="lightblue", fg="black")
label.pack(pady=10)

original_text_box = ScrolledText(root, width=50, height=10, wrap=tk.WORD)
original_text_box.pack()

summarize_button = Button(root, text="Summarize", command=summarize_text, bg="pink", font=("Arial", 14))
summarize_button.pack(pady=10)

label = Label(root, text="Summarized Text", font=("Arial", 14), bg="lightblue", fg="black")
label.pack(pady=10)

summarized_text_box = ScrolledText(root, width=50, height=10, wrap=tk.WORD)
summarized_text_box.pack()

root.mainloop()



```

```python
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu
import matplotlib.pyplot as plt

# Provided text (assuming it as both the reference and generated summary)
text = """In a world inundated with information, it's easy to feel like negativity looms around every corner. From the moment we wake up to the time we close our eyes, we're bombarded with stories of conflict, tragedy, and despair. Whether it's through the news, social media, or everyday conversations, negativity seems to have permeated every aspect of our lives.

It's not hard to see why this is the case. Bad news sells. Sensationalism grabs attention. And in an interconnected world where information spreads at the speed of light, the most shocking and disheartening stories often take center stage. We're constantly reminded of the injustices, the suffering, and the hardships faced by people all around the globe.

Moreover, the human mind has a tendency to dwell on negativity. We're wired to pay more attention to potential threats than to positive events. This evolutionary trait, while once crucial for survival, now leaves us susceptible to focusing on the negative aspects of life rather than the positive ones.

The pervasiveness of negativity can take a toll on our mental and emotional well-being. It breeds fear, anxiety, and hopelessness, leading many to feel overwhelmed and disillusioned. It's easy to lose faith in humanity and to succumb to despair when negativity seems to be the prevailing narrative.

However, it's important to remember that the world is not solely defined by its negativity. For every story of despair, there are countless acts of kindness, resilience, and hope. Behind the headlines are individuals and communities coming together to support one another, to enact positive change, and to create a brighter future.

While it may sometimes feel like we're surrounded by negativity, we have the power to shift the narrative. By seeking out and amplifying stories of hope and compassion, by practicing gratitude and empathy in our own lives, and by actively working towards positive change, we can counteract the pervasive influence of negativity.

Ultimately, the world is what we make of it. Yes, there is darkness and despair, but there is also light and hope. It's up to each of us to choose which narrative we want to contribute to and to strive towards creating a world where positivity prevails."""

# Assuming the provided text is the generated summary and also using it as the reference summary for this example
generated_summary = text
reference_summary = text

# Tokenize the summaries for BLEU calculation
generated_summary_tokens = generated_summary.split()
reference_summary_tokens = [reference_summary.split()]

# Calculate BLEU score
bleu_score = sentence_bleu(reference_summary_tokens, generated_summary_tokens)

# Calculate ROUGE scores
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
rouge_scores = scorer.score(reference_summary, generated_summary)

# Extract ROUGE scores
rouge_1 = rouge_scores['rouge1'].fmeasure
rouge_2 = rouge_scores['rouge2'].fmeasure
rouge_L = rouge_scores['rougeL'].fmeasure

# Plotting the scores
scores = {
    'BLEU': bleu_score,
    'ROUGE-1': rouge_1,
    'ROUGE-2': rouge_2,
    'ROUGE-L': rouge_L
}

# Create the bar plot
fig, ax = plt.subplots()
ax.bar(scores.keys(), scores.values(), color=['blue', 'green', 'red', 'purple'])
ax.set_ylim(0, 1)
ax.set_ylabel('Score')
ax.set_title('Summarization Scores')

# Display the scores above the bars
for i, v in enumerate(scores.values()):
    ax.text(i, v + 0.02, f'{v:.2f}', ha='center', va='bottom')

plt.show()
```
